# Dealing with Outliers

Dependent variable before outlier removal

```python
#Histogram of price before removal of outliers
sns.histplot(house_df['price'],kde=True,bins=20)
plt.title('Histogram of price before removal of outliers')
plt.show()
```

```python
# Creating a copy of the dataframe to remove outliers to see if it works before applying to our real data.
X_test = house_df.copy()

# Isolating numeric variables
numeric_variables = X_test.select_dtypes(include=np.number).columns.to_list()

numeric_variables
```

```python
# Function for checking outliers in a dataframe removing them and creating a new dataframe without outliers
def outliers(house_df):

    # Dictionary for placing no. of outliers and the column name.
    outliers_dict = {}
    standardized_dict = {}

    # For loop for checking outliers in every column in the dataframe.
    for column in house_df.columns:
        # Lower quantile
        Q1 = house_df[column].quantile(0.25)

        # Upper Quantile
        Q3 = house_df[column].quantile(0.75)

        # Finding IQR
        IQR = Q3 - Q1

        # Defining lower and upper bound
        lower_bound = Q1 - (1.5 * IQR)
        upper_bound = Q3 + (1.5 * IQR)

        # Finding outliers
        outliers = house_df[column][(house_df[column] < lower_bound) | (house_df[column] > upper_bound)]
        outliers_dict[column] = 'The number of outliers is', len(outliers)

        # Data without outliers
        standardized_numvotes = house_df[column][(house_df[column] >= lower_bound) & (house_df[column] <= upper_bound)]
        
        # Putting standardized values in a dict
        standardized_dict[column] = standardized_numvotes

    
    standardized_df = pd.DataFrame(standardized_dict)  

    # Results
    return standardized_df
```

```python
house_df=outliers(X_test[numeric_variables])
# Since all values which are not within the range will now be Nan values 
house_df.dropna(inplace=True)

# Histogram of price after removing outliers.
sns.histplot(house_df['price'], kde=True, color='green', bins='auto')
plt.title('Histogram of SalePrices after removing outliers');
```

## Simple linear regression

```python
#Relationship between our target variable 'price' and 'sqft_living'
house_df.plot.scatter(x='sqft_living',y='price');
```

### Baseline model

```python
import statsmodels.api as sm
y = house_df["price"]
X_baseline = house_df[["sqft_living"]]

baseline_model = sm.OLS(y, sm.add_constant(X_baseline))
baseline_results = baseline_model.fit()

print(baseline_results.summary())
```

```python
baseline_results.params
```

- R-squared:measures the proportion of the variance in the dependent variable that is predictable from the independent variable(s). The R-squares is 0.36 indicating approximately 36.2% of the variability in the "price" variable can be explained by the "sqft_living" variable.

- The coefficient estimates show the relationship between the independent variable (sqft_living) and the dependent variable (price).The constant coefficient is 139260.878345 representing the estimated price when sqft-living is zero.The coefficient of sqft_living shows that for each additional sqft_living space ,the model predicts an increase in house price by approximately 171.113805 assuming other variables are held constant.

- The p-values (P>|t|) associated with the coefficients determine their statistical significance. In this case, both the const and sqft_living coefficients have p-values of 0.000, indicating that they are statistically significant.

```python
#fit the simple linear regression model
y = house_df["price"]
X_baseline = house_df["sqft_living"]

# Addinding a constant term to the independent variable matrix
X_baseline = sm.add_constant(X_baseline)

# Creating and fitting the baseline model
baseline_model = sm.OLS(y, X_baseline)
baseline_results = baseline_model.fit()

# Visualize the linearity
plt.scatter(X_baseline['sqft_living'], y, label='Actual')  # Scatter plot of actual values
plt.plot(X_baseline['sqft_living'], baseline_results.predict(), color='blue', label='Predicted')  # Line plot of predicted values
plt.xlabel('sqft_living')
plt.ylabel('Price')
plt.title('Actual vs. Predicted Values')
plt.legend()
plt.show()
```

```python
#checking for homoscedasticity 
#residuals for predicted values, 
fig, ax = plt.subplots()

ax.scatter(house_df["price"], baseline_results.resid)
ax.axhline(y=0, color="black")
ax.set_xlabel("sqft_living")
ax.set_ylabel("residuals"); 
```

- checking for homoscedasticity to conclude the assumptions that the variance of errors (residuals) is constant across all levels of the independent variables.
- Spread of residuals is constant along the predicted values

```python
#visualizing the regression line to show model fit
#Scatter plot of data points
plt.scatter(house_df["sqft_living"], house_df["price"], label="Data points")

# Regression line
x = np.linspace(house_df["sqft_living"].min(), house_df["sqft_living"].max(), 100)
y = baseline_results.params[0] + baseline_results.params[1] * x
plt.plot(x, y, color="black", label="Regression line")

plt.xlabel("sqft_living")
plt.ylabel("Price")
plt.title("Regression Line - Model Fit")
plt.legend()
plt.show()
```

The plot helps to assess the linearity of the relationship between the house price and the square footage of living space.the model shows how well linear regression model fits the data.
